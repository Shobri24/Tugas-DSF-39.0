#**m. Handling Outlier**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk deteksi dan handling outlier dengan IQR
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Visualisasi sebelum handling
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(y=df[column])
    plt.title(f'Before Handling {column}')

    # Handling outlier - clipping
    df[column] = np.where(df[column] > upper_bound, upper_bound,
                         np.where(df[column] < lower_bound, lower_bound, df[column]))

    # Visualisasi setelah handling
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[column])
    plt.title(f'After Handling {column}')
    plt.show()

# Kolom numerik yang perlu di-handle outlier-nya
numeric_cols = ['lead_time', 'adr', 'days_in_waiting_list', 'previous_cancellations']
for col in numeric_cols:
    handle_outliers(data, col)
```

#**n. Encoding variabel**

```python
from sklearn.preprocessing import LabelEncoder

## Persiapan Data
available_cols = data.columns.tolist()

## Label Encoding untuk Variabel Ordinal
ordinal_cols = ['reserved_room_type', 'assigned_room_type', 'arrival_date_month']
ordinal_cols = [col for col in ordinal_cols if col in available_cols]

if ordinal_cols:
    label_encoder = LabelEncoder()
    for col in ordinal_cols:
        data[col] = label_encoder.fit_transform(data[col])
    print("Label Encoding berhasil untuk kolom:", ordinal_cols)
else:
    print("Tidak ada kolom ordinal yang tersedia untuk di-encode")

## One-Hot Encoding untuk Variabel Nominal
nominal_cols = ['hotel', 'meal', 'deposit_type', 'customer_type', 'market_segment']
nominal_cols = [col for col in nominal_cols if col in available_cols]

if nominal_cols:
    data = pd.get_dummies(data, columns=nominal_cols, drop_first=True)
    print("One-Hot Encoding berhasil untuk kolom:", nominal_cols)
    print("Kolom baru yang dibuat:", [col for col in data.columns if any(x in col for x in nominal_cols)])
else:
    print("Tidak ada kolom nominal yang tersedia untuk di-encode")

## Verifikasi Hasil
print("\n5 baris pertama data setelah encoding:")
print(data.head())

print("\nTipe data setelah encoding:")
print(data.dtypes)
```

```python
# ======================
# 1. PERSIAPAN DATA
# ======================

# Set gaya visualisasi
sns.set_theme(style="whitegrid")
plt.style.use('ggplot')
# Memuat data dengan encoding eksplisit
try:
    data = pd.read_csv('h.csv', encoding='utf-8')
except UnicodeDecodeError:
    try:
        data = pd.read_csv('h.csv', encoding='latin1')
    except Exception as e:
        print(f"Gagal memuat file: {e}")
        raise

# Konversi tipe data dengan penanganan error
def convert_to_datetime(series):
    try:
        return pd.to_datetime(series, errors='coerce')
    except Exception as e:
        print(f"Gagal mengkonversi ke datetime: {e}")
        return series

data['reservation_status_date'] = convert_to_datetime(data['reservation_status_date'])
data['is_canceled'] = data['is_canceled'].astype(int)

# Definisi kolom
numeric_cols = ['lead_time', 'adr', 'days_in_waiting_list', 'previous_cancellations',
                'stays_in_week_nights', 'stays_in_weekend_nights', 'children']
categorical_cols = ['hotel', 'customer_type', 'deposit_type', 'market_segment',
                   'arrival_date_month', 'meal']

# ======================
# 2. VISUALISASI DATA
# ======================

# Set gaya global
sns.set_theme(style="whitegrid")
sns.set_theme()

# 2.1 Distribusi Fitur Numerik
def plot_numeric_distributions(data, cols, figsize=(18, 24)):
    plt.figure(figsize=figsize)
    for i, col in enumerate(cols, 1):
        plt.subplot(5, 2, i)
        sns.histplot(data[col], kde=True, bins=30, color='skyblue')
        plt.title(f'Distribusi {col}', fontsize=12)
        plt.xlabel('')
        plt.ylabel('')

    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.suptitle('Distribusi Fitur Numerik', fontsize=16, y=0.99)
    plt.show()

plot_numeric_distributions(data, numeric_cols)

# 2.2 Distribusi Fitur Kategorikal
def plot_categorical_distributions(data, cols, figsize=(18, 24)):
    plt.figure(figsize=figsize)
    for i, col in enumerate(cols, 1):
        plt.subplot(4, 2, i)
        ax = sns.countplot(data=data, x=col, palette='viridis',
                          order=data[col].value_counts().index)
        plt.title(f'Distribusi {col}', fontsize=12)
        plt.xlabel('')


        if len(data[col].unique()) > 5:
            plt.xticks(rotation=45)

        # Tambahkan label nilai untuk 5 kategori teratas
        if i <= 5:  # Hanya untuk beberapa plot pertama agar tidak terlalu padat
            for p in ax.patches:
                ax.annotate(f'{p.get_height():,.0f}',
                           (p.get_x() + p.get_width() / 2., p.get_height()),
                           ha='center', va='center', xytext=(0, 5),
                           textcoords='offset points', fontsize=8)

    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.suptitle('Distribusi Fitur Kategorikal', fontsize=16, y=0.99)
    plt.show()

plot_categorical_distributions(data, categorical_cols)

# 2.3 Tren Pembatalan Bulanan
def plot_monthly_trend(data, date_col='reservation_status_date', target_col='is_canceled'):
    plt.figure(figsize=(16, 8))

    # Pastikan format datetime dan set sebagai index
    monthly_data = data.copy()
    monthly_data[date_col] = pd.to_datetime(monthly_data[date_col])
    monthly_data = monthly_data.set_index(date_col)

    # Resample dan hitung tingkat pembatalan
    cancel_monthly = monthly_data[target_col].resample('M').mean() * 100


    plt.plot(cancel_monthly.index, cancel_monthly.values,
             marker='o', linestyle='-', color='crimson', linewidth=2, markersize=8)
    plt.fill_between(cancel_monthly.index, cancel_monthly.values, alpha=0.2, color='crimson')

    # Formatting
    plt.title('Tren Tingkat Pembatalan Bulanan (%)', fontsize=14)
    plt.xlabel('Tanggal', fontsize=12)
    plt.ylabel('Tingkat Pembatalan (%)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)

    # Rotasi label x-axis
    plt.xticks(rotation=45)

    # Tambahkan garis horizontal pada nilai rata-rata
    mean_rate = cancel_monthly.mean()
    plt.axhline(y=mean_rate, color='navy', linestyle='--',
                label=f'Rata-rata: {mean_rate:.1f}%')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_monthly_trend(data)

# 2.4 Matriks Korelasi
def plot_correlation_matrix(data, cols, figsize=(14, 10)):
    plt.figure(figsize=figsize)

    # Hitung korelasi
    corr_matrix = data[cols + ['is_canceled']].corr()

    # Buat mask untuk segitiga atas
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # Plot heatmap dengan gaya yang lebih baik
    sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f",
                cmap='coolwarm', vmin=-1, vmax=1,
                annot_kws={"size": 10}, cbar_kws={"shrink": 0.8})

    plt.title('Matriks Korelasi Fitur', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

plot_correlation_matrix(data, numeric_cols)

# 2.5 Tingkat Pembatalan per Bulan
def plot_cancellation_by_month(data, month_col='arrival_date_month',
                             target_col='is_canceled', figsize=(14, 7)):
    plt.figure(figsize=figsize)

    # Urutan bulan
    month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                  'July', 'August', 'September', 'October', 'November', 'December']

    # Hitung tingkat pembatalan
    cancel_by_month = data.groupby([month_col, target_col]).size().unstack()
    cancel_by_month = cancel_by_month.reindex(month_order)
    cancel_percentage = cancel_by_month.div(cancel_by_month.sum(axis=1), axis=0) * 100

    # Plot dengan gaya yang lebih baik
    ax = cancel_percentage.plot(kind='bar', stacked=True,
                              color=['lightgreen', 'salmon'],
                              width=0.8, edgecolor='black', linewidth=0.5)

    plt.title('Tingkat Pembatalan per Bulan', fontsize=14)
    plt.ylabel('Persentase (%)', fontsize=12)
    plt.xlabel('Bulan', fontsize=12)

    # Tambahkan label nilai (hanya untuk persentase signifikan)
    for p in ax.patches:
        height = p.get_height()
        if height > 5:
            ax.annotate(f'{height:.1f}%',
                       (p.get_x() + p.get_width()/2, p.get_y() + height/2),
                       ha='center', va='center', fontsize=9,
                       bbox=dict(boxstyle='round,pad=0.2', fc='white', alpha=0.7))

    plt.legend(['Tidak Dibatalkan', 'Dibatalkan'], title='Status',
               bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_cancellation_by_month(data)

# 2.6 Analisis Segmentasi Pelanggan
def plot_customer_segmentation(data, figsize=(18, 12)):
    plt.figure(figsize=figsize)

    # Tipe Pelanggan
    plt.subplot(2, 2, 1)
    sns.countplot(data=data, x='customer_type', hue='is_canceled',
                 palette='Set2', order=data['customer_type'].value_counts().index)
    plt.title('Pembatalan berdasarkan Tipe Pelanggan', fontsize=12)
    plt.xlabel('Tipe Pelanggan', fontsize=10)
    plt.ylabel('Jumlah', fontsize=10)
    plt.legend(title='Dibatalkan', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Tipe Deposit
    plt.subplot(2, 2, 2)
    deposit_cancel = data.groupby('deposit_type')['is_canceled'].mean().sort_values()
    sns.barplot(x=deposit_cancel.index, y=deposit_cancel.values,
               palette='viridis', edgecolor='black')
    plt.title('Tingkat Pembatalan berdasarkan Tipe Deposit', fontsize=12)
    plt.xlabel('Tipe Deposit', fontsize=10)
    plt.ylabel('Tingkat Pembatalan', fontsize=10)
    plt.ylim(0, 1)

    # Lead Time
    plt.subplot(2, 2, 3)
    sns.boxplot(data=data, x='is_canceled', y='lead_time',
               showfliers=False, palette='pastel')
    plt.title('Lead Time vs Pembatalan', fontsize=12)
    plt.xlabel('Dibatalkan', fontsize=10)
    plt.ylabel('Lead Time (hari)', fontsize=10)

    # Permintaan Khusus
    plt.subplot(2, 2, 4)
    sns.countplot(data=data, x='total_of_special_requests',
                 hue='is_canceled', palette='Set3')
    plt.title('Permintaan Khusus vs Pembatalan', fontsize=12)
    plt.xlabel('Jumlah Permintaan Khusus', fontsize=10)
    plt.ylabel('Jumlah', fontsize=10)
    plt.legend(title='Dibatalkan', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()

plot_customer_segmentation(data)

# ======================
# 3. ANALISIS MENDALAM
# ======================

# 3.1 Analisis Segment Pasar
def plot_market_segment_analysis(data, figsize=(12, 6)):
    plt.figure(figsize=figsize)

    market_cancel = data.groupby('market_segment')['is_canceled'].mean().sort_values(ascending=False)

    # Plot dengan gaya yang lebih baik
    ax = sns.barplot(x=market_cancel.index, y=market_cancel.values,
                    palette='rocket', edgecolor='black')

    plt.title('Tingkat Pembatalan berdasarkan Segment Pasar', fontsize=14)
    plt.ylabel('Tingkat Pembatalan', fontsize=12)
    plt.xlabel('Segment Pasar', fontsize=12)

    # Tambahkan label nilai
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}',
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 5),
                   textcoords='offset points', fontsize=10)

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_market_segment_analysis(data)

# 3.2 Analisis Tipe Kamar
def plot_room_type_analysis(data, figsize=(12, 6)):
    plt.figure(figsize=figsize)

    room_cancel = data.groupby('reserved_room_type')['is_canceled'].mean().sort_values()

    # Plot dengan gaya yang lebih baik
    ax = sns.barplot(x=room_cancel.index, y=room_cancel.values,
                    palette='mako', edgecolor='black')

    plt.title('Tingkat Pembatalan berdasarkan Tipe Kamar', fontsize=14)
    plt.ylabel('Tingkat Pembatalan', fontsize=12)
    plt.xlabel('Tipe Kamar', fontsize=12)

    # Tambahkan label nilai
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}',
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 5),
                   textcoords='offset points', fontsize=10)

    plt.tight_layout()
    plt.show()

plot_room_type_analysis(data)

# 3.3 Analisis Faktor Kunci
def plot_key_factors(data, features, figsize=(14, 10)):
    plt.figure(figsize=figsize)

    for i, feature in enumerate(features, 1):
        plt.subplot(2, 3, i)

        if data[feature].dtype == 'object':
            # Untuk fitur kategorikal
            order = data[feature].value_counts().index
            sns.boxplot(data=data, x=feature, y='is_canceled',
                       showfliers=False, order=order)
            plt.xticks(rotation=45)
        else:
            # Untuk fitur numerik
            sns.scatterplot(data=data, x=feature, y='is_canceled',
                          alpha=0.3, color='royalblue')

        plt.title(f'Pengaruh {feature}', fontsize=12)
        plt.xlabel(feature, fontsize=10)
        plt.ylabel('Pembatalan' if i in [1, 4] else '', fontsize=10)

    plt.suptitle('Faktor Kunci yang Mempengaruhi Pembatalan', fontsize=16, y=0.97)
    plt.tight_layout()
    plt.show()

important_features = ['lead_time', 'adr', 'deposit_type', 'customer_type',
                     'total_of_special_requests', 'previous_cancellations']
plot_key_factors(data, important_features)
```
